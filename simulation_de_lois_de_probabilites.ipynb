## 2.1 Introduction

### Exemple Pour la loi normale :

```r
# Spécification d'une seed
set.seed(20242026)

n<-30   #Taille de l'échantillon 30
# Simulation d'une loi normale d'espérance 1 et de variance 4 (écart-type = 2)
ech<-rnorm(n,mean = 0,sd = 2)
# Affichage des 20 premières valeurs de l'échantillon
head(ech,n = 20)
```

```r
## Graphique 1
    # Création de l'histogramme de ech
    hist(ech, freq=F)
    # Ajout de la courbe de densité de a loi normale
    curve(dnorm(x,mean=0,sd=2),from=-6,to=6,ylab="densité",add=T,col="red")
```

```r
## Graphique 2
    # Tracer le graphique de ech
    plot(ecdf(ech))
    # Tracer la fonction de répartition de ech
    curve(pnorm(x,mean=0,sd=2),from=-6,to=6,ylab="Fonction de répartition",add=T,col="red")
```

```r
# Création du quantile de la loi normale d'espérance 1 et d'écart-type = 2
quantile.normale<-qnorm(p=c(0.05,0.1,0.25,0.5,0.75,0.9,0.95),mean = 0,sd = 2)
# Création du quantile de la simulation "ech"
quantile.echantillon<-quantile(x=ech, probs=c(0.05,0.1,0.25,0.5,0.75,0.9,0.95))
# Tableau de comparaison des 2 quantiles
(rbind(quantile.echantillon,quantile.normale))
```

## 2.2 Lois discrètes

### 2.2.1 Loi uniforme discrète

```r
k <- 30  # Création de la taille de l'échantillon
# Simulation d'un échantillon de taille n et de valeurs {1,...,10}.
ech<-sample(1:10,k,replace=T)

# Affichage de la table de propotion de la simulation
table(ech)/k
```

```r
## Graphique 1
    # Création d'une graphique bar
    barplot(table(ech)/k)
    # Trace de la probabilite de la loi uniforme discrete sur {1,...,10}.
    abline(h=1/10,col="red")
```

### 2.2.2 Loi de Bernoulli

```r
n<- 30 # Taille de l'échantillon
p<-0.3  # Paramètre de la loi de Bernoulli
esp<-p # Espérence théorique
var<-p*(1-p) # espérance théorique


ech<-rbinom(n,size = 1,prob = p) # size = 1 correspond à la loi de Bernoulli
table(ech)/n
```

```r
paste("La moyenne de l'échantillon est ",mean(ech),", sachant que l'espérance vaut",esp)
paste("La variance de l'échantillon est ",round(var(ech),2),", sachant que la variance de la loi vaut",p*(1-p))
```

### 2.2.3 Loi de Binomiale

```r
N<- 100 # taille de l'échantillon : nombre d'observations
n<- 10  # premier paramètre de la loi : nombre d'essais
p_1<- 0.8 # deuxième paramètre de la loi : probabilité de succès
p_2<- 0.2 # deuxième paramètre de la loi : probabilité de succès

ech_1<-rbinom(N,n,p_1) # Simulation d'un échantillon pour p=0.8
ech_2<-rbinom(N,n,p_2) # Simulation d'un échantillon pour p=0.2
# Affichage des la table de propotion de la simulation
table(ech_1)/N ; table(ech_2)/N
```

```r
dbinom(x=0:n,size = n,prob = p_1) # Simulation de la densité de la loi binomiale pour p=0.8 
dbinom(x=0:n,size = n,prob = p_2) #  Simulation de la densité de la loi binomiale pour p=0.2
```

```r
N_2<- 10000 # taille de l'échantillon : nombre d'observations
ech_3<-rbinom(N_2,n,p_1) # Simulation d'un échantillon pour p=0.8 et N=1000
ech_4<-rbinom(N_2,n,p_2) # Simulation d'un échantillon pour p=0.2 et N=1000

table(ech_3)/N_2
table(ech_4)/N_2
```

## 2.4 Loi de Poisson

```r
lambda <- 3 # paramètre de la loi de Poisson
n <- 25 # taille de l'échantillon
ech <- rpois(n,lambda) # simulation de la loi de Poisson
barplot(table(ech))
```

```r
# A l’aide de simulation, vérifier que la loi binomiale peut être approchée par la loi de poisson.
#  En pratique, on dit que cette approximation est bonne dès que n>20 et p≤0.1 et np≤5.
n <- 1000
p <- 0.12
lamb <- n*p
binom <- rbinom(n, 1000, p)
poiss <- rpois(n, lamb)

# Créer un data.frame avec les données des deux distributions
df <- data.frame(
  value = c(binom, poiss),
  distribution = c(rep("Binomiale", length(binom)), rep("Poisson", length(poiss)))
)

# Tracer les histogrammes avec ggplot2
ggplot(df, aes(x = value, fill = distribution)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 30) +
  geom_density(alpha = 0.4) +
  labs(title = "Distribution Binomiale et Poisson", x = "Value", y = "Density") +
  theme_minimal()
```

```r
# Tracer les histogrammes avec ggplot2
ggplot(df, aes(x = value, fill = distribution)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 30) +
  geom_density(alpha = 0.4) +
  labs(x = "Value", y = "Density") +
  theme_minimal()
```

```r
# A l’aide de simulation, vérifier que la loi binomiale peut être approchée par la loi de poisson.
#  En pratique, on dit que cette approximation est bonne dès que n>20 et p≤0.1 et np≤5.
n <- 100
lamda_1 <- 2
lamda_2 <- 5
poiss <- rpois(n, lambda = lamda_1)
poiss2 <- rpois(n, lambda = lamda_2)

# faire un graphique histograme avec les 2 distributions lois poissons en utulisant ggplot2 avec la densité en y
library(ggplot2)
df <- data.frame(value = c(poiss, poiss2), distribution = c(rep("lambda = 2", n), rep("lambda = 5", n)))
ggplot(df, aes(x=value, fill=distribution)) + 
  geom_histogram(aes(y=..density..), position="identity", alpha=0.5, bins=30) + 
  geom_density(alpha=0.5)
```

## 2.5 Comparaison des fonctions de répartition empirique et théorique

```r
# Loi uniforme discrete comparaison des fonctions de répartition
k <- 100
ech <- runif(k,1,10)
plot(ecdf(ech))
curve(punif(x,1,10),add=T,col="red")
```

```r
# Loi de Bernoulli comparaison des fonctions de répartition
n <- 100
p <- 0.3
ech <- rbinom(n,size = 1,prob = p)
plot(ecdf(ech))
curve(pbinom(x,1,p),add=T,col="red")
```

```r
# Loi de binomiale comparaison des fonctions de répartition
N <- 1000
n <- 10
p <- 0.3
ech <- rbinom(N,n,p)
plot(ecdf(ech))
curve(pbinom(x,n,p),add=T,col="red")
```

```r
# Loi de Géométrique comparaison des fonctions de répartition
n <- 100
p <- 0.3
ech <- rgeom(n,p)+1
plot(ecdf(ech))
curve(pgeom(x-1,p),add=T,col="red")
```

```r
# Loi de Poisson comparaison des fonctions de répartition
n <- 100
lambda <- 3
ech <- rpois(n,lambda)
plot(ecdf(ech))
curve(ppois(x,lambda),add=T,col="red")
```
