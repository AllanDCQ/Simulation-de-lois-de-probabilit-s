### 2.2.2 Loi de Bernoulli

```r
n<- 30 # Taille de l'échantillon
p<-0.3  # Paramètre de la loi de Bernoulli
esp<-p # Espérence théorique
var<-p*(1-p) # espérance théorique


ech<-rbinom(n,size = 1,prob = p) # size = 1 correspond à la loi de Bernoulli
table(ech)/n
```

```r
paste("La moyenne de l'échantillon est ",mean(ech),", sachant que l'espérance vaut",esp)
paste("La variance de l'échantillon est ",round(var(ech),2),", sachant que la variance de la loi vaut",p*(1-p))
```

### 2.2.3 Loi de Binomiale

```r
N<- 100 # taille de l'échantillon : nombre d'observations
n<- 10  # premier paramètre de la loi : nombre d'essais
p_1<- 0.8 # deuxième paramètre de la loi : probabilité de succès
p_2<- 0.2 # deuxième paramètre de la loi : probabilité de succès

ech_1<-rbinom(N,n,p_1) # Simulation d'un échantillon pour p=0.8
ech_2<-rbinom(N,n,p_2) # Simulation d'un échantillon pour p=0.2
# Affichage des la table de propotion de la simulation
table(ech_1)/N ; table(ech_2)/N
```

## 2.4 Loi de Poisson

```r
lambda <- 3 # paramètre de la loi de Poisson
n <- 25 # taille de l'échantillon
ech <- rpois(n,lambda) # simulation de la loi de Poisson
barplot(table(ech))
```

```r
# A l’aide de simulation, vérifier que la loi binomiale peut être approchée par la loi de poisson.
#  En pratique, on dit que cette approximation est bonne dès que n>20 et p≤0.1 et np≤5.
n <- 1000
p <- 0.12
lamb <- n*p
binom <- rbinom(n, 1000, p)
poiss <- rpois(n, lamb)

# Créer un data.frame avec les données des deux distributions
df <- data.frame(
  value = c(binom, poiss),
  distribution = c(rep("Binomiale", length(binom)), rep("Poisson", length(poiss)))
)

# Tracer les histogrammes avec ggplot2
ggplot(df, aes(x = value, fill = distribution)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 30) +
  geom_density(alpha = 0.4) +
  labs(title = "Distribution Binomiale et Poisson", x = "Value", y = "Density") +
  theme_minimal()
```

```r
# A l’aide de simulation, vérifier que la loi binomiale peut être approchée par la loi de poisson.
#  En pratique, on dit que cette approximation est bonne dès que n>20 et p≤0.1 et np≤5.
n <- 100
lamda_1 <- 2
lamda_2 <- 5
poiss <- rpois(n, lambda = lamda_1)
poiss2 <- rpois(n, lambda = lamda_2)

# faire un graphique histograme avec les 2 distributions lois poissons en utulisant ggplot2 avec la densité en y
library(ggplot2)
df <- data.frame(value = c(poiss, poiss2), distribution = c(rep("lambda = 2", n), rep("lambda = 5", n)))
ggplot(df, aes(x=value, fill=distribution)) + 
  geom_histogram(aes(y=..density..), position="identity", alpha=0.5, bins=30) + 
  geom_density(alpha=0.5)
```

### 3.5 Vérification du théorème 
$\text{Si } \mathbf{X\leadsto\mathcal N(\mu,\sigma^2)} \text{ alors } \mathbf{\dfrac{X-\mu}\sigma\leadsto\mathcal N(0,1)}$

```r
n <- 2000
mu <- 1
sigma <- 4
ech <- rnorm(n, mu, sigma)

# Création du dataframe pour les données simulées
df <- data.frame(x = ech)

# Création du graphique avec ggplot2
p<- ggplot(df, aes(x = (x - mu) / sigma)) +
  # Densité des données simulées
  geom_density(aes(color = "Densité simulée"), fill = "#00BFC4", alpha = 0.5) +
  # Densité théorique de la loi normale
  stat_function(fun = dnorm, aes(color = "Densité théorique"), size = 1.2) +
  scale_color_manual(
    name = "Légende",
    values = c("Densité simulée" = "#00BFC4", "Densité théorique" = "red"),
    labels = c("Densité simulée", "Densité théorique")
  ) +
  labs(x = "Valeurs", y = "Densité", color = "Légende") +
  theme_minimal() +
  theme(legend.position = "right")
p
```

## 5. Lien entre lois normales

Soit $X$ une variable aléatoire distribuée suivant la loi normale d’espérance 0 et d’écart-type 1
On souhaite montrer que les variables $X_1=X+1$, $X_2=X−1$, $X_3=2X$ et $X_4=X/2$ suivent aussi une loi normale. Pour cela on simule un grand nombre de valeurs suivant la loi normale N(0,1) et l’on construit les nouvelles variables. Étudier la distribution de ces nouvelles variables.

```r
set.seed(31)
library(ggplot2)
library(dplyr)
library(gridExtra)

x <- rnorm(10000, 0, 1)
x1 <- x + 1
x2 <- x - 1
x3 <- x * 2
x4 <- x / 2

# Comparaison entre x et x1
df1 <- data.frame(value = c(x, x1), distribution = c(rep("X", length(x)), rep("X+1", length(x1))))
p1 <- ggplot(df1, aes(x = value, fill = distribution)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 30) +
  geom_density(alpha = 0.5)

# Comparaison entre x et x2
df2 <- data.frame(value = c(x, x2), distribution = c(rep("X", length(x)), rep("X-1", length(x2))))
p2 <- ggplot(df2, aes(x = value, fill = distribution)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 30) +
  geom_density(alpha = 0.5)

# Comparaison entre x et x3
df3 <- data.frame(value = c(x, x3), distribution = c(rep("X", length(x)), rep("X*2", length(x3))))
p3 <- ggplot(df3, aes(x = value, fill = distribution)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 30) +
  geom_density(alpha = 0.5)

# Comparaison entre x et x4
df4 <- data.frame(value = c(x, x4), distribution = c(rep("X", length(x)), rep("X/2", length(x4))))
p4 <- ggplot(df4, aes(x = value, fill = distribution)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 30) +
  geom_density(alpha = 0.5)

# Afficher les 4 graphiques sur la même ligne
f<- grid.arrange(p1, p2, p3, p4, ncol = 4)
# enregister le graph f
ggsave("./f.png", f, width = 20, height = 6)
```

```r
moyenneH <- 172
varianceH <- 196

# 1. quelle est la probabilité que X soit inférieure à 160cm ? Pr(X < 160)
question1 <- pnorm(160, moyenneH, sqrt(varianceH))
print(question1)
```

## 9. Loi Exponentielle

```r
x <- seq(0, 10, 0.1)
lambda <- 0.5
f <- lambda * exp(-lambda * x)

# Tracer la fonction densité avec ggplot
df <- data.frame(x = x, f = f)
ggplot(df, aes(x = x, y = f)) +
  geom_line(color = "blue") +
  labs(title = "Fonction densité de la loi exponentielle", x = "x", y = "f(x)") +
  theme_minimal()

# vérifier l'ntégration à 1
integrate(function(x) lambda * exp(-lambda * x), 0, Inf)
```
